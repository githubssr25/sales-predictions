```python
# Project 1 - Part 3 (Core)
# We will continue to work on your sales prediction project. 
# The goal of this is to help you understand the data as it relates to the  
# products and outlets that play crucial roles in increasing sales.

# For Part3, your task is to complete any statistical analyses that might help you understand,
#  explain, or model with your data. This must include at least one of each:

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
# corr = df.corr()

filename = '/content/sales_predictions (3).csv'



df = pd.read_csv(filename)
df.head(20)

corr = df.corr()



```

```python
df.describe()
# summary statistics 
```

```python
mean_MRP = df.groupby(['Item_Type'])['Item_Outlet_Sales'].mean()

print (mean_MRP)

mean_weight = df.groupby(['Item_Type'])['Item_MRP'].mean()

print(mean_weight)

mean_fat = df.groupby(['Item_Fat_Content'])['Item_Outlet_Sales'].mean()

print(mean_fat)

by_year = df.groupby(['Outlet_Establishment_Year'])['Item_Outlet_Sales'].mean()

print(by_year)

by_size = df.groupby(['Outlet_Size'])['Item_Outlet_Sales'].mean()

print(by_size)

item_visible = df.groupby(['Item_Type'])['Item_Visibility'].median()

print(item_visible)

outlet_type_sales = df.groupby(['Outlet_Type'])['Item_Outlet_Sales'].mean()

print(outlet_type_sales)
```

```python
plt.figure(figsize = (12, 5))

plt.barh(mean_MRP.index, mean_MRP)

plt.xlabel('MRP')
plt.title('Average MRP of Each Type of Item');
```

```python
plt.figure(figsize = (12, 5))

plt.barh(by_size.index, by_size)

plt.xlabel('Item Outlet Sales')
plt.title('Sales By Size Of Outlet');
```

```python
plt.figure(figsize = (12, 5))

plt.barh(item_visible.index, item_visible)

plt.xlabel('Average Visibility Rating')
plt.title('Average Visibility Rating Of Items In Stores');
```

```python
plt.figure(figsize = (12, 5))

plt.barh(outlet_type_sales.index, outlet_type_sales)

plt.xlabel('Average Sales By Type of Supermarket Or Grocery Store')
plt.title('Average Sales');
```

```python
MRP = df['Item_MRP']

weight = df['Item_Weight']

fig = plt.gcf()

fig.set_size_inches(17, 10)

sns.scatterplot(x = weight, y = MRP, data = df)

plt.xlabel('Premeium ')
plt.ylabel('Median Total Complaints')
plt.title('How Premeium Varies With Number Of Complaints');
```
