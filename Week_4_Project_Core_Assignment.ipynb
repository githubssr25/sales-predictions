```python
sales_by_fat = df.groupby(['Item_Fat_Content'])['Item_Outlet_Sales'].mean()

sales_by_type = df.groupby(['Item_Type'])['Item_Outlet_Sales'].mean()

fat_by_year = df.groupby(['Item_Fat_Content'])['Outlet_Establishment_Year'].mean()

sales_by_year = df.groupby(['Outlet_Establishment_Year'])['Item_Outlet_Sales'].mean()

weight_by_year = df.groupby(['Outlet_Establishment_Year'])['Item_Weight'].mean()

sales_year_index = list(sales_by_year.index)

sales_year_value = list(sales_by_year.values)

weight_year_index = list(weight_by_year.index)

weight_year_value = list(weight_by_year.values)
```

```python
plt.style.use('seaborn')
plt.plot(sales_year_index, sales_year_value, c = (0, 1, 1), marker = '.', markersize = 10)

plt.plot(sales_year_index, sales_year_value, c= 'k')
plt.xlim(left=1984,right=2010)
plt.ylim(bottom=0,top=2600)
plt.xlabel('Year', fontsize = 16)
plt.ylabel('Sales', fontsize = 16)
plt.title('Sales Over the Years', fontsize = 16)

# this will graph the item outlet sales vs the establishment year. So how does the sales change at newer outlets as they are established over time. 
```
